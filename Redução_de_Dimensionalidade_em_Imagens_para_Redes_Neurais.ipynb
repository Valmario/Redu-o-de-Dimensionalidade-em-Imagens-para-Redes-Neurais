```python
# Importando bibliotecas necessárias
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

# Upload da imagem
from google.colab import files
uploaded = files.upload()

# Carregando a imagem enviada
image_path = list(uploaded.keys())[0]
imagem_original = cv2.imread(image_path)

# Exibindo a imagem original
print("Imagem Original:")
cv2_imshow(imagem_original)

# Convertendo para tons de cinza
imagem_cinza = cv2.cvtColor(imagem_original, cv2.COLOR_BGR2GRAY)
print("Imagem em Tons de Cinza:")
cv2_imshow(imagem_cinza)

# Aplicando a binarização
_, imagem_binaria = cv2.threshold(imagem_cinza, 127, 255, cv2.THRESH_BINARY)
print("Imagem Binarizada (Preto e Branco):")
cv2_imshow(imagem_binaria)

# Salvando as imagens geradas
cv2.imwrite("imagem_cinza.png", imagem_cinza)
cv2.imwrite("imagem_binaria.png", imagem_binaria)
```
